### A funciton for temperature histogram for a specific region

```python
def temp_hist(band = 'LST_Day_1km',date = '2021-06-01', state_fp = '46'):
    counties = ee.FeatureCollection('TIGER/2018/Counties').filter(ee.Filter.eq('STATEFP', state_fp))
    region  = counties.union()
    img = ee.ImageCollection("MODIS/061/MOD11A1").filterDate(ee.Date(date), ee.Date(date).advance(1,'day')).first()
    arrt = img.select(band).multiply(.02).subtract(273).reduceRegion(ee.Reducer.histogram(),region).getInfo()
    dft = pd.DataFrame(data = {'temp': arrt[band]['bucketMeans'], 'n_occur':arrt[band]['histogram']})
    fig, ax = plt.subplots(figsize=(20, 6))
    ax.grid(lw=0.2)
    ax.scatter(dft['temp'],dft['n_occur'],c='black', alpha=0.7, label='temprature_histogram')
    return plt.show()
```

### Growing degree days calculation 

```python
def GDD_calculator_on_img(baseT = 5, state_fp = '46',temperature_band = 'LST_Day_1km'):
    counties = ee.FeatureCollection('TIGER/2018/Counties').filter(ee.Filter.eq('STATEFP', state_fp))
    state = counties.union() 
    def gdd_on_img(img):
        gdd = img.select(temperature_band).multiply(.02).subtract(273).subtract(baseT)
        mask = gdd.gte(0)
        gdd = gdd.updateMask(mask) 
        mean_gdd = gdd.reduceRegion(ee.Reducer.mean(), state)
        gdd = gdd.set('mean_gdd',mean_gdd)
        gdd_SD = gdd.clip(state)
        return gdd_SD
    return gdd_on_img
```

```python
def cum_GDD(gddlis):
    def none_to_0(x):
        if x == None:
            return 0
        else:
            return x
    l = [i for i in map(none_to_0,gddlis)]
    return np.cumsum(l)  
```

```python
def GDD_mean_calc(date = '2020-01-01',duration_month = 12,cum = True,baseT = 5, state_fp = '46',temperature_band = 'LST_Day_1km', img_col = "MODIS/061/MOD11A1"):
    """
    this function computes mean gdd over a state for a date and duration
    """
    c = ee.ImageCollection(img_col).filterDate(ee.Date(date), ee.Date(date).advance(duration_month,'month'))
    f = GDD_calculator_on_img(baseT = baseT, state_fp = state_fp,temperature_band = temperature_band)
    c = c.map(f)
    l= c.aggregate_array('mean_gdd').getInfo()
    gddlist = [d['LST_Day_1km'] for d in l]
    if cum:
        return cum_GDD(gddlist)
    else:
        return gddlist
```

```python
def GDD_img_calc(date = '2020-01-01',duration_month = 12,cum = True,baseT = 5, state_fp = '46',temperature_band = 'LST_Day_1km', img_col = "MODIS/061/MOD11A1"):

    c = ee.ImageCollection(img_col).filterDate(ee.Date(date), ee.Date(date).advance(duration_month,'month'))
    f = GDD_calculator_on_img(baseT = baseT, state_fp = state_fp,temperature_band = temperature_band)
    c = c.map(f)
    return c
```
